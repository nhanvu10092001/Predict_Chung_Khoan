# Neural Network


```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

training_X, test_X, training_y, test_y = train_test_split(Original_feature, Original_label, test_size = 0.15)

scaler = MinMaxScaler()
training_Scalered = scaler.fit_transform(training_X)
test_scalered = scaler.transform(test_X)

import sklearn.externals 
import joblib

import pickle
scalerfile = 'scaler.sav'
pickle.dump(scaler, open(scalerfile, 'wb'))
```

```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset
import torch.optim as optim
from torch.utils.data import DataLoader
from tqdm.auto import tqdm
from IPython.display import clear_output
```

```python
class Regression(nn.Module):
    def __init__(self, input_feature):
        super(Regression, self).__init__()

        self.layer_0 = nn.Linear(input_feature, 100)
        self.layer_1 = nn.Linear(100, 80)
        self.layer_2 = nn.Linear(80, 60)
        self.layer_3 = nn.Linear(60, 20)
        self.layer_4 = nn.Linear(20, 5)
    def forward(self, x):
        
        output = self.layer_0(x)
        output = self.layer_1(output)
        output = self.layer_2(output)
        output = self.layer_3(output)
        output = self.layer_4(output)
        return output


class FPTDataset(Dataset):
    def __init__(self, array, label):
        super(FPTDataset, self).__init__()

        self.data = array.astype(np.float32)
        self.label = label.astype(np.float32)
    def __len__(self):
        return self.data.shape[0]

    def __getitem__(self, index):
        return self.data[index, :], self.label[index, :]
        
```

```python
model = Regression(180)

loss_function = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.00001)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=20)

train_data_set = FPTDataset(array = training_Scalered, label =  training_y)
test_data_set = FPTDataset(array = test_scalered, label =  test_y)





```

```python

train_dataloader = DataLoader(train_data_set, batch_size=2, shuffle=True, drop_last = True)
test_dataloader = DataLoader(test_data_set, batch_size=1, shuffle=False)
```

```python
loss_init = torch.inf
bad_couter = 0 # count how many time test loss < current loss
max_bad = 15 # maximum times reach bad count until break training

for i in range (0,10):
    progress_bar_train = tqdm(train_dataloader)

    model.train()
    train_loss = 0
    for sample in progress_bar_train:

        feature, label = sample
        
        output =  model(feature)
    
        loss = loss_function(output, label)
        loss.backward()
        optimizer.step()
        model.zero_grad()
        train_loss += loss


    train_loss = train_loss / len(train_dataloader)
    print(f"train loss {train_loss}")  
    model.eval()
    with torch.no_grad():

        
        progress_bar_test = tqdm(test_dataloader)
        test_loss = 0

        for sample in progress_bar_test:
            
            feature, label = sample        
            output =  model(feature)
            test_loss +=loss_function(output, label)
        test_loss = test_loss / len(test_dataloader)
        scheduler.step(test_loss)
        print(f"test loss {test_loss}")
        
        if loss_init > test_loss:
            loss_init = test_loss
            bad_couter = 0
        else:
            bad_couter += 1
        if bad_couter > max_bad:
            break
    clear_output(wait=True)
```

```python
model.eval()
with torch.no_grad():
    progress_bar_test = tqdm(test_dataloader)
    test_loss = 0
    labels = []
    predictions = []
    for sample in progress_bar_test:
        feature, label = sample        
        output =  model(feature)
        test_loss +=loss_function(output, label)
        labels.append(str(label.tolist()))
        predictions.append(str(output.detach().cpu().tolist()))
    test_loss = test_loss / len(test_dataloader)
    print(f"test loss {test_loss}")
```
